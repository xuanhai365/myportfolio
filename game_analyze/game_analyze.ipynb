**Distribution of numerical features**

```python
plt.hist(df.Year, bins = 20)
plt.title('Number of Games released by Year')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (8, 8))

ax[0][0].pie([sum(df.NA_Sales <= 1),sum(df.NA_Sales > 1)], autopct='%1.1f%%', pctdistance=1.25)
ax[0][0].set_title('North America')

ax[0][1].pie([sum(df.EU_Sales <= 1),sum(df.EU_Sales > 1)], autopct='%1.1f%%', pctdistance=1.25)
ax[0][1].set_title('Europe')

ax[1][0].pie([sum(df.JP_Sales <= 1),sum(df.JP_Sales > 1)], autopct='%1.1f%%', pctdistance=1.25)
ax[1][0].set_title('Japan')

ax[1][1].pie([sum(df.Other_Sales <= 1),sum(df.Other_Sales > 1)], autopct='%1.1f%%', pctdistance=1.25)
ax[1][1].set_title('Other')

fig.legend(['Sales <= 1 mil.', 'Sales > 1 mil.'])
fig.suptitle('Percentage % of Game Sales in different Region')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (8, 8), sharex = True, sharey = True)

ax[0][0].hist(df.NA_Sales[df.NA_Sales <= 1])
ax[0][0].set_title('North America')

ax[0][1].hist(df.EU_Sales[df.EU_Sales <= 1])
ax[0][1].set_title('Europe')

ax[1][0].hist(df.JP_Sales[df.JP_Sales <= 1])
ax[1][0].set_title('Japan')

ax[1][1].hist(df.Other_Sales[df.Other_Sales <= 1])
ax[1][1].set_title('Other')

fig.suptitle('Distribution of Game Sales less than 1 mil. in different Region')
fig.supxlabel('Number of Games')
fig.supylabel('Revenue (Milions $)')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))

ax[0].pie([sum(df.Global_Sales <= 5),sum(df.Global_Sales > 5)], autopct='%1.1f%%', labeldistance=.6, pctdistance=1.25)
ax[0].set_title('Percentage % of Game Sales across the world')
ax[0].legend(['Sales <= 5 mil.', 'Sales > 5 mil.'])

ax[1].hist(df.Global_Sales[df.Global_Sales <= 5])
ax[1].set_title('Distribution of Game Sales less than 5 mil. across the world')
ax[1].set_xlabel('Number of Games')
ax[1].set_ylabel('Revenue (Milions $)')

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10, 4))
sales_by_year = df.Global_Sales.groupby(df.Year).sum()

ax[0].plot(sales_by_year.keys() , sales_by_year)
ax[0].set_title('Total Sales Revenue of Game Industry by Year')

ax[1].hist(df.Year)
ax[1].set_title('Total Number of Games released by Year')
plt.show()
```

```python
sales_by_year = df.Global_Sales.groupby(df.Year).sum()
game_by_year = df.Name.groupby(df.Year).count()
average_revenue = sales_by_year/game_by_year

plt.plot(average_revenue.keys() , average_revenue)
plt.title('Average Revenue of Game Industry by Year')
plt.xlabel('Year')
plt.ylabel('Revenue (Milions $)')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (30, 8))
sales_by_publish = df.Global_Sales.groupby(df.Publisher).sum()
sales_by_publish = sales_by_publish.sort_values().tail(10)

ax[0].barh(sales_by_publish.keys(), sales_by_publish)
ax[0].set_ylabel('Publisher')
ax[0].set_xlabel('Revenue (Milions $)')

sales_by_genre = df.Global_Sales.groupby(df.Genre).sum()
sales_by_genre = sales_by_genre.sort_values().tail(10)

ax[1].barh(sales_by_genre.keys(), sales_by_genre)
ax[1].set_ylabel('Genre')
ax[1].set_xlabel('Revenue (Milions $)')

sales_by_platform = df.Global_Sales.groupby(df.Platform).sum()
sales_by_platform = sales_by_platform.sort_values().tail(10)

ax[2].barh(sales_by_platform.keys(), sales_by_platform)
ax[2].set_ylabel('Platform')
ax[2].set_xlabel('Revenue (Milions $)')

fig.suptitle('Most Revenue from Publisher, Genre and Platform')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 4))
popular_genre = df.Name.groupby(df.Genre).count()
popular_genre = popular_genre.sort_values().tail(10)

ax[0].barh(popular_genre.keys(), popular_genre)
ax[0].set_ylabel('Genre')
ax[0].set_xlabel('Number of Games')

popular_platform = df.Name.groupby(df.Platform).count()
popular_platform = popular_platform.sort_values().tail(10)

ax[1].barh(popular_platform.keys(), popular_platform)
ax[1].set_ylabel('Platform')
ax[1].set_xlabel('Number of Games')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 4))

ax[0].plot(df.Name[df.Genre == 'Action'].groupby(df.Year).count(), label = 'Action')
ax[0].plot(df.Name[df.Genre == 'Sports'].groupby(df.Year).count(), label = 'Sports')
ax[0].plot(df.Name[df.Genre == 'Misc'].groupby(df.Year).count(), label = 'Misc')
ax[0].set_ylabel('Number of Games')

ax[1].plot(df.Name[df.Platform == 'DS'].groupby(df.Year).count(), label = 'DS')
ax[1].plot(df.Name[df.Platform == 'PS2'].groupby(df.Year).count(), label = 'PS2')
ax[1].plot(df.Name[df.Platform == 'PS3'].groupby(df.Year).count(), label = 'PS3')
ax[1].set_ylabel('Number of Games')

ax[0].legend()
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (12, 8))

total_sale = df.Global_Sales.sum()
nintendo_sale = df.Global_Sales[df.Publisher == 'Nintendo'].sum()
EA_sale = df.Global_Sales[df.Publisher == 'Electronic Arts'].sum()
active_sale = df.Global_Sales[df.Publisher == 'Activision'].sum()

ax[0][0].pie([nintendo_sale, EA_sale, active_sale, total_sale - nintendo_sale - EA_sale - active_sale],
        labels = ['Nintendo', 'EA', 'Activision', 'Other'], autopct='%1.1f%%')
ax[0][0].set_title('Percentage % of top 3 largest Publisher in Game Industry')

ax[0][1].plot(df.Global_Sales[df.Publisher == 'Nintendo'].groupby(df.Year).sum(), label = 'Nintendo Sales')
ax[0][1].plot(df.Global_Sales[df.Publisher == 'Electronic Arts'].groupby(df.Year).sum(), label = 'Electronic Arts Sales')
ax[0][1].plot(df.Global_Sales[df.Publisher == 'Activision'].groupby(df.Year).sum(), label = 'Activision Sales')
ax[0][1].set_title('Sale Revenue of top 3 largest Publisher by Year')
ax[0][1].set_ylabel('Revenue (Milions $)')
ax[0][1].legend()

ax[1][0].plot(df.Name[df.Publisher == 'Nintendo'].groupby(df.Year).count(), label = 'Nintendo Games')
ax[1][0].plot(df.Name[df.Publisher == 'Electronic Arts'].groupby(df.Year).count(), label = 'Electronic Arts Games')
ax[1][0].plot(df.Name[df.Publisher == 'Activision'].groupby(df.Year).count(), label = 'Activision Games')
ax[1][0].set_title('Number of Games of top 3 largest Publisher by Year')
ax[1][0].set_ylabel('Revenue (Milions $)')
ax[1][0].legend()

average_nintendo = df.Global_Sales[df.Publisher == 'Nintendo'].groupby(df.Year).sum()/df.Name[df.Publisher == 'Nintendo'].groupby(df.Year).count()
average_EA = df.Global_Sales[df.Publisher == 'Electronic Arts'].groupby(df.Year).sum()/df.Name[df.Publisher == 'Electronic Arts'].groupby(df.Year).count()
average_active = df.Global_Sales[df.Publisher == 'Activision'].groupby(df.Year).sum()/df.Name[df.Publisher == 'Activision'].groupby(df.Year).count()
ax[1][1].plot(average_nintendo, label = 'Nintendo')
ax[1][1].plot(average_EA, label = 'Electronic Arts')
ax[1][1].plot(average_active, label = 'Activision')
ax[1][1].set_title('Average Revenue of top 3 largest Publisher by Year')
ax[1][1].set_ylabel('Revenue (Milions $)')
ax[1][1].legend()
plt.show()
```

```python
NAsales = df.NA_Sales.groupby(df.Publisher).sum()
EUsales = df.EU_Sales.groupby(df.Publisher).sum()
JPsales = df.JP_Sales.groupby(df.Publisher).sum()
Osales = df.Other_Sales.groupby(df.Publisher).sum()

index = ['Nintendo', 'Electronic Arts', 'Activision']
columns = ['NA Sales','EU Sales','JP Sales','Other Sales']
data = pd.DataFrame([[NAsales['Nintendo'],EUsales['Nintendo'],JPsales['Nintendo'], Osales['Nintendo']],
                    [NAsales['Electronic Arts'],EUsales['Electronic Arts'],JPsales['Electronic Arts'], Osales['Electronic Arts']],
                    [NAsales['Activision'],EUsales['Activision'],JPsales['Activision'], Osales['Activision']]], index = index, columns = columns)
data.plot.bar(rot = 0)

plt.ylabel('Revenue (Milions $)')
plt.show()
```
